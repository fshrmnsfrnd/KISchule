```python
import numpy
from sklearn.datasets import fetch_california_housing
dataset = fetch_california_housing()
x = dataset.data[:, :]
y = dataset.target
```

# Polynomielle Regression
## Aufgabe 1
Laden Sie erneut das california_housing Dataset aus der letzten Übung. Zerlegen Sie erneut das Dataset in x und y Daten und machen Sie anschließend wiederrum die Aufteilung in 70% Trainingsdaten und 30% Testdaten.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import MinMaxScaler

#Normalisierung der Daten
scaler= MinMaxScaler()
x= scaler.fit_transform(x)
print(x.shape[1])
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

```

## Aufgabe 2
- Wenden Sie nun die Polynomielle Regression auf das Dataset an. Verwenden Sie zuerst ein Polynom 2.Grades.
- Geben Sie im nächsten Schritt den Shape der original Daten sowie der transformierten Daten aus.
- Wenden Sie dann auf die transformierten Daten die Lineare Regression an.
- Vergleichen Sie den R2-Score (Erinnerung ursprünglicher R2-Score: 0.59)

```python
degree = 2
pf = PolynomialFeatures(degree=degree) #Erzeugen des Objekts mit dem Grad des Polynom
pf.fit(x_train)
x_train_transformed = pf.transform(x_train) #Transformieren der Daten
x_test_transformed = pf.transform(x_test)
print(x_train_transformed.shape[1])
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(x_train_transformed, y_train)
r2_score = regr.score(x_test_transformed, y_test)

print(f"R2-Score: {r2_score}")
```

Können Sie den R2-Score noch optimieren? Testen Sie ein Polynom höheren Grades. Nutzen Sie dazu GridSearch (Achtung! Rechenzeit steigt an)
