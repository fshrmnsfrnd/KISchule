# Vorbereitung Dataset für die Übungsaufgabe

```python
import numpy as np
np.random.seed(42)

from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()
x = dataset.data
y = dataset.target

```

## Aufgabe 1
Wir nutzen nochmals das Brustkrebsdataset um die Normalisierung der Daten zu testen. Bestimmen Sie mit GridSearch die besten Parameter für den Random Forest. Führen Sie die Suche einmal mit normalisierten Daten und einmal ohne normalisierte Daten aus (Standard Scaler). Vergleichen Sie den Score und Speichern Sie das beste Model ab.
Nutzen Sie dazu folgende Parameter:
- #Anzahl der Bäume= 100
- Kriterium: gini, entropy
- Maximale Tiefe: 2-8
- Mindesanzahl an Datenpunkten: 10-20
- Kreuzvalidierung: 10
- n_jobs=-1
- Aufteilung des Datasets: train:70%, test:30%

Rechenzeit: ca. 50s
  

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)

clf = RandomForestClassifier(min_samples_leaf=2, max_features="log2")
parameters = {
    "criterion": ["gini", "entropy"],
    "max_depth": [2, 3, 4, 5, 6, 7, 8],
    "min_samples_split": [2, 4, 6],
    "min_samples_leaf": [1, 2],
    "max_features": ["sqrt", "log2"]
}
grid_cv = GridSearchCV(clf, parameters, cv=10, n_jobs=-1)
grid_cv.fit(x_train, y_train)
print("Best Params:", grid_cv.best_params_)
print(grid_cv.cv_results_)
```

```python
from sklearn.preprocessing import StandardScaler 

scaler=StandardScaler()  
scaler.fit(x)

x_transformed=scaler.transform(x) 
x_train, x_test, y_train, y_test = train_test_split(x_transformed, y, test_size=0.3)
clf = RandomForestClassifier(min_samples_leaf=2, max_features="log2")
parameters = {
    "criterion": ["gini", "entropy"],
    "max_depth": [2, 3, 4, 5, 6, 7, 8],
    "min_samples_split": [2, 4, 6],
    "min_samples_leaf": [1, 2],
    "max_features": ["sqrt", "log2"]
}
grid_cv = GridSearchCV(clf, parameters, cv=10, n_jobs=-1)
grid_cv.fit(x_train, y_train)
print("Best Params:", grid_cv.best_params_)
print(grid_cv.cv_results_)
```
