# Dataset 

```python
import numpy as np

np.random.seed(42)
from sklearn import datasets
```

```python
wine = datasets.load_wine()

x = wine.data[:, :]

y = wine.target

class_names = wine.target_names
discription = wine.DESCR
```

# Aufgabe 1
Wir beginnen in der ersten Aufgabe mit der Entwicklung eines Neuronalen Netzes für das bekannte Wein Dataset.

- Bestimmen Sie für die Ausgabeschicht die richtige Größe (y).
- Führen Sie anschließend wieder die Aufteilung nach Training und Testdaten durch (70% Training).

```python
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
print(y)
y_cat=to_categorical(y,num_classes=3) #denke optional
print(y_cat.shape)
x_train, x_test, y_train, y_test = train_test_split(x, y_cat, test_size=0.30)
```

# Aufgabe 2
Erstellen Sie ein Neuronales Netzwerk für dieses Klassifikationsproblem. 
- Nutzen Sie mindestens 2 verdeckte Schichten. 
- Verwenden Sie jewils 450 Neuronen und als Aktivierungsfunktion (Relu/bzw. softmax).
- Geben Sie danach eine Zusammenfassung des Modells aus und Trainieren Sie dieses Abschließend.

```python
from keras import layers
from tensorflow.keras.layers import Activation
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import SGD

model = Sequential()
model.add(layers.Input(shape=(x.shape[1],),name='Eingabeschicht'))
model.add(Dense(units=450,name='Erste_verdeckte_Schicht'))
model.add(Activation("relu",name='Aktivierungsfunktion_Schicht1'))
model.add(Dense(units=450,name='Zweite_verdeckte_Schicht'))
model.add(Activation("relu",name='Aktivierungsfunktion_Schicht2'))
model.add(Dense(units=3,name='Ausgabeschicht'))
model.add(Activation("softmax",name='Aktivierungsfunktion_der_Ausgabeschicht'))
```

```python
model.summary()
```

## Aufgabe 3
- Compilieren Sie das Neuronale Netz mit Hilfe einer geeigneten Fehlerfunktion, einem geeigneten Optimierer und einer sinnvollen Metrik.
- Trainieren Sie anschließend das Modell mit 30 Epochen. Nutzen Sie dazu ebenfalls die Testdaten.

```python
model.compile(
	loss="categorical_crossentropy", optimizer=SGD(), metrics=["accuracy"]
)
```

```python
model.fit(x=x_train, y=y_train, epochs=30, validation_data=(x_test,y_test))
```

# Aufgabe 4
Für das zweite Neuronale Netz, verwenden wir das Calfornia Housing Dataset (Hauspreise). Entsprechend ein Regressionsproblem.
- Wie groß müsste die Ausgabeschicht hier sein?
- Führen Sie wieder die Aufteilung nach Training und Testdaten durch (70% Training).

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 

dataset = fetch_california_housing(data_home="/cal_housing_py3pkz")
x = dataset.data
y = dataset.target

scaler=StandardScaler()  
scaler.fit(x)
x_transformed=scaler.transform(x) 
x_train, x_test, y_train, y_test = train_test_split(x_transformed, y, test_size=0.3)

```

# Aufgabe 5
- Erstellen Sie ein Neuronales Netzwerk für dieses Regressionsproblem. 
- Nutzen Sie mindestens 2 bis 3 verdeckte Schichten. 
- Verwenden Sie für die Anzahl der Neuronen eine Pyramiden Struktur und als Aktivierungsfunktion Relu. 
- Geben Sie danach eine Zusammenfassung des Modells aus.
- Trainieren Sie dieses Modell abschließend. Welche Größe muss die Ausgabeschicht besitzen?
- Für die Fehlerfunktion (loss) verwenden Sie bitte den mittleren quadratischen Fehler (MSE) (tf.keras.losses.mse) und für die Metrik den MAE.
- Verwenden Sie 10 Epochen

```python
from keras import layers
from tensorflow.keras.layers import Activation
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import SGD

model = Sequential()
model.add(layers.Input(shape=(x.shape[1],),name='Eingabeschicht'))
model.add(Dense(units=256,name='Erste_verdeckte_Schicht'))
model.add(Activation("relu",name='Aktivierungsfunktion_Schicht1'))
model.add(Dense(units=128,name='Zweite_verdeckte_Schicht'))
model.add(Activation("relu",name='Aktivierungsfunktion_Schicht2'))
model.add(Dense(units=1,name='Ausgabeschicht'))
model.add(Activation("softmax",name='Aktivierungsfunktion_der_Ausgabeschicht'))
```

```python
model.summary()
```

```python
import tensorflow as tf
model.compile(
	loss=tf.keras.losses.mse, optimizer=SGD(), metrics=["MAE"]
)
```

```python
model.fit(x=x_train, y=y_train, epochs=10, validation_data=(x_test,y_test))
```
