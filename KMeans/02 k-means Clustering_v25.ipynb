## Übungsaufgabe zum kmeans Clustering
### Aufgabe 1

Gegeben ist das nachfolgende Wine Dataset. Von diesem sind in dieser Übung nur die Feature Daten (x-Dataset) bekannt.
Erzeugen Sie mit Matplotlib Punktwolken aus den Featuere kombination (Feature1 und jeweils allen anderen Featuren)
Versuchen Sie aus diesen die mögliche Clusteranzahl zu bestimmen. 

```python
import matplotlib.pyplot as plt
from sklearn import datasets

wine = datasets.load_wine()
x = wine.data
y = wine.target

allFeatureCompinations = []
i = 0
for i in range (len(x)):
    curData = x[i]
    curFeatureCombination = []
    curFeatureCombination.append(curData[0])
    j = 1
    for j in range (len(curData)):
        curFeatureCombination.append(curData[j])
    allFeatureCompinations.append(curFeatureCombination)
import numpy as np
allFeatureCompinations = np.array(allFeatureCompinations)
plt.scatter(allFeatureCompinations[:, 0], allFeatureCompinations[:, 1], c='blue')
plt.plot(x[:, 0], x[:, 1], 'ro')
```

## Aufgabe 2
Aus den Charts erkennt man nicht wirklich eine sinnvolle Gruppierung von Datenpunkten. Deshalb folgt nun die Analyse mit Hilfe des Elbow_Chart.
Erzeugen Sie mit Hilfe der bereit gestellten Funktion (Elbow_Chart_Funktion) einen Elbow_Chart. Bestimmen Sie hierheraus die sinnvollste Clusteranzahl.

```python
from sklearn.cluster import KMeans
from Elbow_Chart_Funktion import create_elbow_chart 
kmeans = KMeans(n_clusters=2, random_state=0, n_init="auto")
kmeans.fit(x, y)
labels = kmeans.labels_ #Bestimmte Klassen für jeden Punkt
centers = kmeans.cluster_centers_ #Cluster Mittelpunkte
create_elbow_chart(x, 10)
```
