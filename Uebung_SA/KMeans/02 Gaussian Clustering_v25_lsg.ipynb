### Aufgabe 2 Gaussian Mixture Algorithmus

In dieser Aufgabe sollen Sie die Ergebnisse des kmeans Clustering Algorithmus (Cluster Zentren) für die Initalisierung des Gaussian Mixture Algorithmus verwenden. Nutzen Sie die gleiche Anzahl an Clustern wie sie beim letzten Mal für das Wine-Dataset bestimmt haben. Nutzen Sie 100 Iterationen für die Initalisierung.

Bestimmen Sie danach die Labels (y-Dataset für das GMM) und vergleichen Sie diese mit den Labels, die Sie mit dem kmeans Algorithmus bestimmt haben. Was fällt Ihnen dabei auf?


```python
from sklearn import datasets

wine = datasets.load_wine()
X_original = wine.data
```

```python
# Daten müssen vor der Verwendung skaliert werden, das Prinizip wird nach der Schulaufgabe erklärt
from sklearn.preprocessing import StandardScaler

scale=StandardScaler()
scale.fit(X_original)
X=scale.transform(X_original)
```

```python
import numpy as np
from sklearn.mixture import GaussianMixture
# Extrahiere die K-Means Clusterzentren und speichere sie in einer Variablen
loaded_kmeans_centers = loaded_kmeans.cluster_centers_
n_clusters = 3
# K-Means Ergebnisse für GMM vorbereiten
kmeans_means = loaded_kmeans_centers

# GMM mit K-Means Initialisierung
gmm_kmeans_init = GaussianMixture(
    n_components=n_clusters,    # Anzahl der Cluster
    means_init=loaded_kmeans_centers,    # Initialisierung der Mittelpunkte
    max_iter=100                # Anzahl der durchzuführenden Initialisierungen (Iterationen). die besten Ergebnisse werden beibehalten
)
# Fitte das GMM Modell und erhalte die Clusterzuordnungen
gmm_kmeans_init.fit(X)
#  Vorhersage der Clusterzuordnungen
gmm_kmeans_labels = gmm_kmeans_init.predict(X)
print(kmeans_means)
print(gmm_kmeans_init.means_)
#y=gmm_kmeans_init.predict(X)
#Vergleich mit den echten Labels
print(wine.target)
print(gmm_kmeans_labels)
```
