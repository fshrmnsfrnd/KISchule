# Lineare Regression

Nachfolgend Imporiteren wir das Dataset, den Algorithmus (bzw. die Klasse LineareRegression).

```python
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
```

Das Dataset umfasst ca. 20.000 Datenpunkte. Nachfolgend wir das Dataset auf den PC geladen. Dies kann je nach Internetverbindung etwas dauern.

```python
dataset = fetch_california_housing()
print(dataset.DESCR)
```

### Aufgabe 1
Wenden Sie die Lineare Regression auf das California Housing dataset an. Teilen Sie das Dataset, wie gewohnt in Trainings- und Testdaten. Bestimmen Sie anschlie√üend den R2-Score und bewerten Sie diesen.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_train, x_test, y_train, y_test = train_test_split(dataset.data, dataset.target, test_size=0.2, random_state=42)

regr = LinearRegression()
regr.fit(x_train, y_train)
r2_score = regr.score(x_test, y_test)
print(f"Coef:\n{regr.coef_}")
print(f"Intercept: {regr.intercept_}")
print(f"R2-Score: {r2_score}")
```
