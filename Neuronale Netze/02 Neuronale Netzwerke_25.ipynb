# Dataset 

```python
import numpy as np

np.random.seed(42)
from sklearn import datasets
```

```python
wine = datasets.load_wine()

x = wine.data[:, :]

y = wine.target

class_names = wine.target_names
discription = wine.DESCR
```

# Aufgabe 1
Wir beginnen in der ersten Aufgabe mit der Entwicklung eines Neuronalen Netzes fГјr das bekannte Wein Dataset.

- Bestimmen Sie fГјr die Ausgabeschicht die richtige GrГ¶Гҹe (y).
- FГјhren Sie anschlieГҹend wieder die Aufteilung nach Training und Testdaten durch (70% Training).

```python
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
y_cat=to_categorical(y,num_classes=3)
print(y)

from sklearn.preprocessing import StandardScaler

x = StandardScaler().fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(x, y_cat, test_size=0.30)
```

# Aufgabe 2
Erstellen Sie ein Neuronales Netzwerk fГјr dieses Klassifikationsproblem. 
- Nutzen Sie mindestens 2 verdeckte Schichten. 
- Verwenden Sie jewils 450 Neuronen und als Aktivierungsfunktion (Relu/bzw. softmax).
- Geben Sie danach eine Zusammenfassung des Modells aus und Trainieren Sie dieses AbschlieГҹend.

```python
import numpy as np
np.random.seed(42)
import tensorflow as tf
tf.random.set_seed(42)      #Auch in Tensorflow kГ¶nnen wir einen Beginn 
                            #fГјr die Zufallszahlen festlegen
from tensorflow import keras
from sklearn.datasets import load_digits        #Import Dataset
from keras.layers import Activation     #Import Modul fГјr die Aktivierungsfunktion
from keras.layers import Dense          #Anzahl der Dichte der "Neuronen"
from keras.models import Sequential     #Modul zum Erzeugen des Neuronalen Netzes
from keras.optimizers import SGD        #Import Modul Stochastic Gradient Decent
from keras.layers import Input

model=Sequential()
model.add(Input(shape=(x.shape[1],)))               #Eingabeschicht
model.add(Dense(units=450, name='Hidden_Layer1'))
model.add(Activation("relu", name='Aktivierungsfunktion_Hidden_Layer1'))
model.add(Dense(units=450, name='Hidden_Layer2'))
model.add(Activation("relu", name='Aktivierungsfunktion_Hidden_Layer2'))
model.add(Dense(units=450, name='Hidden_Layer3'))
model.add(Activation("relu", name='Aktivierungsfunktion_Hidden_Layer3'))
model.add(Dense(units=3))                          #Ausgabeschicht
model.add(Activation("softmax"))
```

```python
model.summary()
```

## Aufgabe 3
- Compilieren Sie das Neuronale Netz mit Hilfe einer geeigneten Fehlerfunktion, einem geeigneten Optimierer und einer sinnvollen Metrik.
- Trainieren Sie anschlieГҹend das Modell mit 30 Epochen. Nutzen Sie dazu ebenfalls die Testdaten.

```python
model.compile(
    loss="categorical_crossentropy", optimizer=SGD(), metrics=["accuracy"]
)
```

```python
model.fit(x=x_train, y=y_train, epochs=100, validation_data=(x_test,y_test))
```

```python
import numpy as np
np.random.seed(42)
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
dataset = fetch_california_housing()
x = dataset.data[:, :]
y = dataset.target
```

```python
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
print(y)

from sklearn.preprocessing import StandardScaler

x = StandardScaler().fit_transform(x)
```

# Aufgabe 4
FГјr das zweite Neuronale Netz, verwenden wir das Calfornia Housing Dataset (Hauspreise). Entsprechend ein Regressionsproblem.
- Wie groГҹ mГјsste die Ausgabeschicht hier sein?
- FГјhren Sie wieder die Aufteilung nach Training und Testdaten durch (70% Training).

```python
print(x.shape)
print(y.shape)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30)
```

# Aufgabe 5
- Erstellen Sie ein Neuronales Netzwerk fГјr dieses Regressionsproblem. 
- Nutzen Sie mindestens 2 bis 3 verdeckte Schichten. 
- Verwenden Sie fГјr die Anzahl der Neuronen eine Pyramiden Struktur und als Aktivierungsfunktion Relu. 
- Geben Sie danach eine Zusammenfassung des Modells aus.
- Trainieren Sie dieses Modell abschlieГҹend. Welche GrГ¶Гҹe muss die Ausgabeschicht besitzen?
- FГјr die Fehlerfunktion (loss) verwenden Sie bitte den mittleren quadratischen Fehler (MSE) (tf.keras.losses.mse) und fГјr die Metrik den MAE.
- Verwenden Sie 10 Epochen

```python
model=Sequential()
model.add(Input(shape=(x.shape[1],)))               #Eingabeschicht
model.add(Dense(units=512, name='Hidden_Layer1'))
model.add(Activation("relu", name='Aktivierungsfunktion_Hidden_Layer1'))
model.add(Dense(units=256, name='Hidden_Layer2'))
model.add(Activation("relu", name='Aktivierungsfunktion_Hidden_Layer2'))
model.add(Dense(units=128, name='Hidden_Layer3'))
model.add(Activation("relu", name='Aktivierungsfunktion_Hidden_Layer3'))
model.add(Dense(units=1))                          #Ausgabeschicht
model.add(Activation("relu"))
```

```python
model.summary()
```

```python
model.compile(
    loss=tf.keras.losses.mse, optimizer=SGD(), metrics=['mae']
)
```

```python
model.fit(x=x_train, y=y_train, epochs=15, validation_data=(x_test,y_test))
```
