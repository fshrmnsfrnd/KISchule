```python
#Mich wundert dass keine Normalisierung gefragt ist, ich hoffe ich hab nix überlesen
#Hätte man es doch machen sollen hier als kommentar wie es gehen würde:
#from sklearn.preprocessing import MinMaxScaler
#scaler = MinMaxScaler()
#scaler.fit(x)
#x_transformed = scaler.transform(x)
#Das müsste man nach den imports und vor train_test_split machen
#Und natürlich in train_test_split nicht x sondern x_transformed verwenden

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#Linear
regr = LinearRegression(n_jobs=-1)
regr.fit(x_train, y_train)
r2_score_linear = regr.score(x_test, y_test)

print(f"Linear R2-Score: {r2_score_linear}")

#Polynomell
pipe=Pipeline(steps=[('pf',PolynomialFeatures()),('reg',LinearRegression(n_jobs=-1))])

params={'pf__degree':[2,3,4]}
grid = GridSearchCV(pipe,params ,cv=5, n_jobs=-1)
grid.fit(x_train, y_train)
bestclf = grid.best_estimator_

bestPipe = Pipeline(steps=[('pf',PolynomialFeatures()),('reg',LinearRegression())])
bestPipe.set_params(pf__degree=2)
bestPipe.fit(x_train, y_train)
r2_score_polynomell = bestPipe.score(x_test, y_test)
print("Polynomell R2 Score: ",r2_score_polynomell)

#Linear R2-Score: 0.8127304254351552
#Polynomell R2 Score:  0.8124424641610775

import pickle
with open('save_model.pkl','wb')as file: 
	pickle.dump(regr,file)
```

## Aufgabe 3 (4P)
Erstellen Sie ein Neuronales Netz mit folgenden Parametern:

Neuronale Netze:
- Eingabeschicht zum Dataset aus Aufgabe 1
- 1 Schichten mit je 64 Neuronen
- Fehlerfunktion: Geeignete für Regression
- Optimierer: Adam
- Metrik: Geeignete für Regression
- Anzahl Epochen: max 10
- Ausgabeschicht
Die Rechenzeit sollte bei unter ca. 1 Minute betragen.

```python
import numpy as np
np.random.seed(42)
import tensorflow as tf
tf.random.set_seed(42)
from keras.layers import Activation
from keras.layers import Dense
from keras.models import Sequential
from keras import layers

model = Sequential()
model.add(layers.Input(shape=(x.shape[1],),name='Eingabeschicht'))
model.add(Dense(units=64,name='Erste_verdeckte_Schicht'))
model.add(Activation("relu",name='Aktivierungsfunktion_Schicht1'))
model.add(Dense(units=1,name='Ausgabeschicht'))
model.add(Activation("softmax",name='Aktivierungsfunktion_der_Ausgabeschicht'))
#model.summary()
model.compile(
	loss=tf.keras.losses.mse, optimizer="Adam", metrics=["MAE"]
)
model.fit(x=x_train, y=y_train, epochs=10, validation_data=(x_test,y_test))
```
