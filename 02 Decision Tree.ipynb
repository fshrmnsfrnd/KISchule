# Decision Tree
Für Übung zu Entscheidungsbäumen verwenden wir erneut das Wine-Dataset. Wir wollen erneut ein Klassifikationsproblem lösen!

```python
import numpy as np
np.random.seed(42)

from sklearn.datasets import load_wine
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
dataset = load_wine()
x = dataset.data
y = dataset.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

## Aufgabe 1 
Trainieren Sie Entscheidungsbäume für ein Klassifikationsproblem. Nutzen Sie zum Testen folgende Parameter:
- gini und entropy

- Minimale Zahl für die Teilung: 2, 4, 6
- Minimale Anzahl übriger Datenwerte: 1, 2
- Die Maximale Anzahl an Features die für den besten Split geprüft werden soll: sqrt, log2

Nutzen Sie zudem für die Crossvalidation den Wert 10 und führen Sie soviele Trainings wie möglich gleichzeitig aus. Bestimmen Sie anschließend die Parameter für das beste Modell. Trainieren Sie abschließend mit den besten Parametern ein neues Modell und lassen Sie sich die Accuracy aus geben.

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(
    criterion="gini", max_depth=4, min_samples_leaf=2, max_features="log2"
)
clf.fit(x_train, y_train)
```

```python
from sklearn.tree import plot_tree
from matplotlib import pyplot as plt
_ = plot_tree(clf)
```
